````python
''' Building a system to evaluate a student's performance based on their test scores.
    The system calculates their average, checks if they passed, determines rank eligibility,
    and evaluates logical conditions based on their performance. '''

# Input data
total_marks = 450  # Total marks obtained (int)
max_marks = 500  # Maximum possible marks (int)
passing_percentage = 40  # Passing percentage (int)
top_rank_percentage = 90  # Percentage required for top rank (int)
extra_credit = 10  # Extra credit for bonus questions (int)

# Mathematical Operations
average_marks = total_marks / 5  # Calculate average marks (division)
percentage_score = (total_marks + extra_credit) / max_marks * 100  # Calculate percentage (+, /, *)
remaining_marks = max_marks - total_marks  # Marks needed to achieve maximum marks (-)
bonus_applied = extra_credit // 2  # Floor division for bonus split (//)
remainder_bonus = extra_credit % 2  # Remainder after bonus division (%)
double_score = total_marks ** 2  # Demonstrate exponentiation (**)

print(f"Average Marks: {average_marks}")
print(f"Percentage Score: {percentage_score:.2f}%")
print(f"Remaining Marks for Full Score: {remaining_marks}")
print(f"Bonus Applied (Floor Division): {bonus_applied}")
print(f"Remaining Bonus Points: {remainder_bonus}")
print(f"Double Score (Exponential Example): {double_score}")

# Comparison Operators
passed = percentage_score >= passing_percentage  # Check if passed (>=)
is_top_rank = percentage_score >= top_rank_percentage  # Check top rank eligibility (>=)
failed = percentage_score < passing_percentage  # Check failure (<)
perfect_score = total_marks == max_marks  # Check if perfect score (==)
needs_improvement = total_marks != max_marks  # Check if needs improvement (!=)

print(f"Passed the Exam? {passed}")
print(f"Eligible for Top Rank? {is_top_rank}")
print(f"Failed the Exam? {failed}")
print(f"Perfect Score? {perfect_score}")
print(f"Needs Improvement? {needs_improvement}")

# Logical Operations
is_consistent = passed and not failed  # Passed and not failed
eligible_for_bonus = passed or percentage_score >= 35  # Passed or near pass
data_type_check = type(total_marks) is int  # Use 'is' to check type of variable

if is_consistent:
    print("Student has shown consistent performance.")
if eligible_for_bonus:
    print("Student is eligible for bonus.")
print(f"Is the 'total_marks' variable an integer? {data_type_check}")
````

# Why Python Does Not Have `i++` or `++i`?

* Python's designers intentionally avoided the use of ++ and -- to keep the language simple and readable. These operators in other languages can sometimes lead to confusion, especially in cases where the increment/decrement is used within expressions.

01. Equivalent Operations in Python:

  Pre-Increment (Simulated)
  ```python
  i = 5
  i += 1  # Pre-increment: incrementing before use
  print(i)  # Output: 6
  ```

  Post-Increment (Simulated)
  ```python
  i = 5
  current_value = i  # Post-increment: use the value first
  i += 1  # Then increment the value
  print(current_value)  # Output: 5 (the value before increment)
  print(i)  # Output: 6 (the value after increment)
  ```

  Why This Difference?

  Python does not support i++ or ++i to keep the language clean and readable. These operators in other languages can be more prone to misuse, especially when used within complex expressions. In Python, the goal is simplicity and clarity.

  `Pre-increment`: Increment the value first, then use it.
  
  `Post-increment`: Use the value first, then increment it.

# Global and local Variables
* **Global Variable**: A global variable is like the public library: everyone in your program can access it, no matter where they are. It’s open to the whole world—so use it wisely!

* **Local Variable**: A local variable is like the snack you hide in your desk drawer: only you (within your function) know about it, and once you're done, it disappears—out of sight, out of mind!

````python
# Global variable
total_balance = 1000

def check_balance():
    # Reading the global variable without 'global' keyword
    print(f"Your current balance is ${total_balance}")

check_balance()
````

````python
# Global variable
total_balance = 1000

def deposit(amount):
    global total_balance  # Access thvariable global
    total_balance += amount  # Add to the total balance
    print(f"Deposited ${amount}. Total Balance: ${total_balance}")

def withdraw(amount):
    global total_balance  # Access the global variable
    if total_balance >= amount:
        total_balance -= amount  # Deduct from the total balance
        print(f"Withdrew ${amount}. Total Balance: ${total_balance}")
    else:
        print("Insufficient balance!")

def check_balance():
    # Local variable for formatting
    balance_message = f"Your current balance is ${total_balance}"  # Local variable
    print(balance_message)

# Using the functions
deposit(500)
withdraw(300)
check_balance()
````

# String Slicing

- String slicing in Python is the process of extracting a portion of a string using its indices, with the format string[start:end:step].

- "String slicing is like making a sandwich: you decide where to cut the bread (start and end), and if you’re feeling fancy, you can even skip every other slice (step)!" 🍞👉🍞

- String slicing allows you to extract parts of a string using the format:
  `string[start:end:step]`

    `start`: The starting index (inclusive).

    `end`: The ending index (exclusive).

    `step`: The interval for slicing.










````python
# String Slicing (includes starting index discludes ending index)
# Slice always returns a new copy
# Valid for all sequence data types ﴾i.e list, tuples,strings﴿

name = "Kenexai India"
print(name[0]) # Accessing the first element
print(name[-1]) # Accessing the last element
print(name[0:3]) # Slicing from index 0 to 3 (exclusive)
````

````python
print(name[:3]) # Slicing from the beginning to index 3 (exclusive)
print(name[1:]) # Slicing from index 1 to the end
print(name[:]) # Full slice (entire string)
````

````python
print(name[::1]) # Starting from the first, select every element
print(name[::-1]) # Starting from the last, select every element (reverse the string)
print(name[::2]) # Starting from the first, select every second element
````

````python
# 1. Basic Slicing

text = "Hello, World!"
substring = text[0:5]  # Slices characters from index 0 to 4
print(substring)
````

````python
# Tuple (immutable)
my_tuple = (1, 2, 3, "Python")
print(my_tuple[2]) # Accessing the third element of the tuple

# List (mutable)
my_list = [10, 20, 30, "Python"]
my_list[1] = 25 # Modifying the second element
print(my_list) # List after modification

# Set: Collection of Unordered and Unique Items
my_set = {1, 2, 3, 3, 4} # Duplicates will be removed
print(my_set) # Output will show unique and unordered items

# Dictionary: Key-value pair, No explicit ordering
my_dict = {"name": "John", "age": 25, "job":"unemployed"}
print(my_dict["name"]) # Accessing the value using the key

# Generic Collections: Can contain objects of any type
generic_list = [42, "hello", 3.14, [1, 2, 3]] # A list containing different types
print(generic_list[3]) # Accessing the nested list

# Nested Collections: Collections within collections
nested_dict = {"details": {"name": "Alice", "age": 30}, "scores": [95, 85, 90]}
print(nested_dict["details"]["name"]) # Accessing nested dictionary value
print(nested_dict["scores"][1]) # Accessing the second item in the nested list
````

````python
#List Manipulation

# Push(Append): "Adding an item to the end of the list, like tossing your shoes into the closet!"
# Pop: "Removing the last item from the list, like grabbing that last cookie from the jar!"
# Peek: "Looking at the last item in the list without removing it, like peeking over the top of the cookie jar!"

# Initialize an empty list
stack = []

# Push elements to the list (like adding items to a stack)
stack.append(10)  # Push 10
stack.append(20)  # Push 20
stack.append(30)  # Push 30
print("After pushes, stack:", stack)

# Peek the last element (check the top item without removing it)
peek_value = stack[-1]  # Peek the last item
print("Peek: The last item in the stack is:", peek_value)

# Pop the last element (remove and get the last item)
popped_value = stack.pop()  # Pop the last item
print("After popping, popped value:", popped_value)
print("Current stack after pop:", stack)

# Pop another element
stack.pop()  # Pop the next element
print("Stack after another pop:", stack)

````

````python
#Collection Operator

# Membership Operator: `in`
my_list = [1, 2, 3, 4]
print(3 in my_list) # Check if 3 is in the list
print(2 not in my_list) # Check if 5 is in the list

# Concatenation: `+`
list1 = [1, 2, 3]
list2 = [4, 5, 6]
print(list1 + list2) # Concatenate two lists

string1 = "Hello"
string2 = " World"
print(string1 + string2) # Concatenate two strings

# Repetition: `*`
my_list = [1, 2, 3]
print(my_list * 2) # Repeat the list twice

my_string = "Hi"
print(my_string * 3) # Repeat the string three times

# Comparison Operators: `<`, `>`, `==`
list1 = [1, 2, 3]
list2 = [1, 2, 4]
print(list1 == list2) # Compare two lists for equality
print(list1 < list2)  # Compare lexicographically

# Membership with Dictionary
my_dict = {"name": "Alice", "age": 30}
print("name" in my_dict) # Check if 'name' is a key
print("Alice" in my_dict.values()) # Check if 'Alice' is a value

# Slice Operator: `[:]`
my_tuple = (10, 20, 30, 40)
print(my_tuple[1:3]) # Slice the tuple from index 1 to 3 (exclusive), Output: (20, 30)

# Length Operator: `len()`
my_set = {1, 2, 3, 4}
print(len(my_set)) # Get the number of elements in the set
````

# Set Operator
* **Union (| or .union())** : Combines two sets and returns a new set with all unique elements from both.
* **Intersection (& or .intersection())** : Returns a new set with only the elements that are common to both sets.
* **Difference (- or .difference())** : Returns a new set with elements in the first set but not in the second.
* **Symmetric Differenc**e (^ or .symmetric_difference()) : Returns a new set with elements in either set, but not in both.
```python
set1 = {1, 2, 3}
set2 = {2, 3, 4}
print(set1 | set2) # Union of set1 and set2, Output: {1, 2, 3, 4}
print(set1.union(set2)) # Using union() method, Output: {1, 2, 3, 4}
print(set1 & set2) # Intersection of set1 and set2, Output: {2, 3}
print(set1.intersection(set2)) # Using intersection() method, Output: {2, 3}
print(set1 - set2) # Elements in set1 but not in set2, Output: {1}
print(set1.difference(set2)) # Using difference() method, Output: {1}
print(set1 ^ set2) # Elements in set1 or set2 but not in both, Output: {1, 4}
print(set1.symmetric_difference(set2)) # Using symmetric_difference() method, Output: {1, 4}
print(set1 <= set2) # Check if set1 is a subset of set2, Output: False
print(set1.issubset(set2)) # Using issubset() method, Output: False
print(set1 >= set2) # Check if set1 is a superset of set2, Output: False
print(set1.issuperset(set2)) # Using issuperset() method, Output: False
print(set1.isdisjoint(set2)) # Check if set1 and set2 have no elements in common, Output: False
set1.add(5) # Add an element to set1
print(set1) # Output: {1, 2, 3, 5}
print(set1) # Output: {1, 3, 5}
```

````python
# Set of registered user emails
registered_users = {
    "user1@example.com",
    "user2@example.com",
    "user2@example.com",
    "user3@example.com",
    "user4@example.com"
}

# Set of users who have unsubscribed
unsubscribed_users = {
    "user2@example.com",
    "user4@example.com"
}

# Use set operations to find subscribed users
subscribed_users = registered_users - unsubscribed_users  # Set difference

print("Subscribed Users (Emails to Send):")
print(subscribed_users)

# Additional Set Operations
all_users = subscribed_users | unsubscribed_users  # Union of both sets
print("\nAll Users (Combined List):")
print(all_users)

common_users = registered_users & unsubscribed_users  # Intersection (users in both sets)
print("\nCommon Users (Registered and Unsubscribed):")
print(common_users)

exclusive_users = registered_users ^ unsubscribed_users  # Symmetric difference
print("\nExclusive Users (Only in One of the Sets):")
print(exclusive_users)

# Check if a specific email is still subscribed
email_to_check = "user3@example.com"
if email_to_check in subscribed_users:
    print(f"\n{email_to_check} is still subscribed.")
else:
    print(f"\n{email_to_check} is not subscribed.")
````

**Some commonly used methods of a dictionary**
* **clear()**: Removes all items from the dictionary, making it empty
 `my_dict.clear()`

 `new_dict = my_dict.copy()`
* **get()**: Retrieves the value associated with the specified key (returns None if the key doesn't exist)
`value = my_dict.get('key')`
* **items()**: Returns a view object that displays a list of dictionary's key-value tuple pairs
 `items = my_dict.items()`
* **keys()**: Returns a view object that displays a list of all the keys in the dictionary
 `keys = my_dict.keys()`
* **pop()**: Removes and returns the value associated with the specified key
`value = my_dict.pop('key')`
* **popitem()**: Removes and returns a random key-value pair from the dictionary
 `item = my_dict.popitem()`
* **setdefault()**: Returns the value of the specified key, and if it doesn't exist, inserts the key with the provided default value
`value = my_dict.setdefault('key', 'default')`
* **update()**: Updates the dictionary with the elements from another dictionary or iterable of key-value pairs
`my_dict.update({'key': 'new_value'})`
* **values()**: Returns a view object that displays a list of all the values in the dictionary
`values = my_dict.values()`

**Use Case**: ETL Process - Data Cleaning and Aggregation

Imagine you're working with a dataset representing sales transactions, and you need to clean up the data and calculate the total sales for a particular product.

See the example below


````python
# Sample dataset: List of dictionaries representing sales transactions
transactions = [
    {"transaction_id": 1, "product_id": 101, "amount": 200, "date": "2025-01-10"},
    {"transaction_id": 2, "product_id": 102, "amount": 300, "date": "2025-01-10"},
    {"transaction_id": 3, "product_id": 101, "amount": None, "date": "2025-01-11"},  # Missing amount
    {"transaction_id": 4, "product_id": 103, "amount": 150, "date": "2025-01-11"},
    {"transaction_id": 5, "product_id": 101, "amount": 250, "date": "2025-01-12"},
]

# Data Cleaning and Aggregation
product_sales = {}

# Loop through each transaction to clean data and aggregate sales
for transaction in transactions:
    # Skip transactions with missing data
    if transaction["amount"] is None:
        print(f"Skipping transaction {transaction['transaction_id']} due to missing amount.")
        continue  # Skip to the next iteration

    product_id = transaction["product_id"]
    amount = transaction["amount"]

    # Add or update total sales for the product
    if product_id not in product_sales:
        product_sales[product_id] = amount
    else:
        product_sales[product_id] += amount

# Output the aggregated sales for each product
print("\nAggregated Sales per Product:")
for product_id, total_sales in product_sales.items():
    print(f"Product ID: {product_id}, Total Sales: ${total_sales}")
````

````python
# While loop
correct_password = "password123"
max_attempts = 3
attempts = 0

while attempts < max_attempts:
    user_input = input("Enter the password: ")

    if user_input == correct_password:
        print("Success! You've entered the correct password.")
        break
    else:
        attempts += 1
        remaining_attempts = max_attempts - attempts
        if remaining_attempts > 0:
            print(f"Incorrect password. You have {remaining_attempts} attempt(s) remaining.")
        else:
            print("Maximum attempts exceeded. You are now locked out.")
````

Use Case: Batch Processing for Large Data

Imagine you're working with a large dataset stored in a file, and you need to process the data in chunks (batch processing). You can simulate this with a while loop.

See the example below



````python
# Sample dataset (a large list simulating data rows)
data = [
    {"transaction_id": i, "product_id": i % 5, "amount": 100 + i}
    for i in range(1, 101)  # Simulate 100 rows of data
]

# Simulating batch processing
batch_size = 20  # Process 20 records per batch
processed_data = []
start_index = 0

# While loop to simulate batch processing
while start_index < len(data):
    # Extract a batch from the dataset
    end_index = start_index + batch_size
    batch = data[start_index:end_index]

    # Process the batch (e.g., clean, transform, or aggregate)
    total_sales = sum(record["amount"] for record in batch)
    print(f"Processing batch from {start_index + 1} to {end_index}. Total sales in this batch: ${total_sales}")

    # Store the processed data (e.g., save to a new list or a database)
    processed_data.extend(batch)

    # Update the start index for the next batch
    start_index = end_index

# Output final result (simulated)
print("\nTotal records processed:", len(processed_data))
````

# Structural Pattern Matching
Structural pattern matching is a programming technique that allows complex data structures to be applied in a concise and readable way in comparisons

```python
def greeting(language):
    chosen_language = language.capitalize()
    match chosen_language:
        case 'English':
            print('Hello')
        case 'German':
            print('Hallo!')
        case _:
            print(f"Hello, I don't know {chosen_language}!")
            
greeting('english')
greeting('french')
```

````python
def get_status_code_description(code):
    match code:
        case 200:
            return "OK"
        case 404:
            return "Not Found"
        case 500:
            return "Internal Server Error"
        case _:
            return "Unknown Status Code"

print(get_status_code_description(200))
````

````python
# Sample data formats (represented as dictionaries)
data_json = {"type": "json", "content": '{"name": "John", "age": 30}'}
data_csv = {"type": "csv", "content": "name,age\nJohn,30"}
data_db = {"type": "db", "content": {"name": "John", "age": 30}}

# Data processing functions for each format
def process_json(content):
    print(f"Processing JSON: {content}")

def process_csv(content):
    print(f"Processing CSV: {content}")

def process_db(content):
    print(f"Processing Database Record: {content}")

# Data processing function using pattern matching
def process_data(data):
    match data:
        case {"type": "json", "content": content}:  # Match data with type 'json'
            process_json(content)
        case {"type": "csv", "content": content}:  # Match data with type 'csv'
            process_csv(content)
        case {"type": "db", "content": content}:   # Match data with type 'db'
            process_db(content)
        case _:  # Default case for unsupported types
            print("Unsupported data format")

# Test with different data formats
print("Test Case 1: JSON Data")
process_data(data_json)

print("\nTest Case 2: CSV Data")
process_data(data_csv)

print("\nTest Case 3: Database Data")
process_data(data_db)

print("\nTest Case 4: Unsupported Format")
process_data({"type": "xml", "content": "<name>John</name>"})
````

````python
# Arbitrary Number of Parameters
def arithmetic_mean(first, *values):
    """ This function calculates the arithmetic mean of a non-empty
        arbitrary number of numerical values """
    return (first + sum(values)) / (1 + len(values))

print(arithmetic_mean(45,32,89,78))
print(arithmetic_mean(8989.8,78787.78,3453,78778.73))
print(arithmetic_mean(45,32))
print(arithmetic_mean(45))
````

````python
# Function inside a Function (Nested Function)
def calculate_invoice(price, tax_rate, discount_code=None):
    # Function to calculate tax
    def calculate_tax(price, tax_rate):
        return price * tax_rate / 100

    # Function to apply discount
    def apply_discount(price, discount_code):
        if discount_code == "DISCOUNT10":
            return price * 0.10
        return 0

    # Calculate tax
    tax = calculate_tax(price, tax_rate)

    # Apply discount if available
    discount = apply_discount(price, discount_code)

    # Final total after tax and discount
    total = price + tax - discount
    return total

# Example usage
invoice_amount = calculate_invoice(100, 5, "DISCOUNT10")
print(f"Total Invoice Amount: ${invoice_amount:.2f}")
````

````python
# Function that returns another function
def currency_converter(rate):
    # This function returns another function to convert the amount
    def convert(amount):
        return amount * rate
    return convert

# Create a USD to EUR converter (e.g., rate = 0.85 EUR per USD)
usd_to_eur = currency_converter(0.85)

# Convert $100 USD to EUR
amount_in_eur = usd_to_eur(100)
print(f"$100 USD is {amount_in_eur} EUR")

# Create a EUR to GBP converter (e.g., rate = 0.88 GBP per EUR)
eur_to_gbp = currency_converter(0.88)

# Convert 100 EUR to GBP
amount_in_gbp = eur_to_gbp(100)
print(f"100 EUR is {amount_in_gbp} GBP")
````

# Exception Handling
* An error that occurs during the execution of a program, disrupting its normal flow. Instead of letting the program crash, Python allows you to handle these errors gracefully using exception handling techniques.
* Handling exceptions is like saying, `'I know something might go wrong, but I've got this!'` Instead of your program crashing like a bad joke falling flat, you catch the error and keep the show running.


**Key Terms:**
* **try**: "Hey Python, try this risky business."
* **except**: "Caught you! Here's how to fix that mess."
* **finally**: "Cleanup on aisle 5! I'm running no matter what."

image.png



```python
try:
    # Attempt something risky
    result = 10 / 0
except ZeroDivisionError:
    # Handle the error gracefully
    print("Oops! Can't divide by zero, not even in Python.")
finally:
    # Clean-up, no matter what
    print("Math class dismissed!")
```

````python
def int_input(prompt):
    while True:
        try:
            age = int(input(prompt))
            if age < 0:
                raise ValueError("Age cannot be negative!")
            return age
        except ValueError as e:
            print(e)


age = int_input("Please enter your age: ")
print(f"Your age is {age}")
````

````python
import csv

# Function to read and process user data from a CSV file
def process_user_data(file_path):
    # try:
        # Open the CSV file
        with open(file_path, mode="r") as file:
            reader = csv.DictReader(file)

            # Process each row in the CSV file
            for row in reader:
                # Simulate processing (e.g., check if 'age' is a valid integer)
                try:
                    age = int(row['age'])  # Convert age to integer
                    print(f"Processing user: {row['name']}, Age: {age}")
                except ValueError:
                    print(f"Error: Invalid age value for user {row['name']}. Skipping.")
    # except FileNotFoundError:
    #     print(f"Error: The file '{file_path}' was not found.")

file_path = 'user_data.csv'  # Path to the CSV file
process_user_data(file_path)
````
